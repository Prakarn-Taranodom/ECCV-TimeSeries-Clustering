##T-test

```python


def t_test_mean_greater_than_zero(data):
    """Performs a one-sample t-test to check if the mean is greater than zero."""
    t_stat, p_value = stats.ttest_1samp(data, 0, alternative='greater')
    result = "yes" if p_value < 0.05 else "no"
    return t_stat, p_value, result

def test_by_dataset(df):
    """Performs t-tests for each dataset to check if the mean difference is greater than zero."""
    results = []
    for dataset in df['Dataset'].unique():
        subset = df[df['Dataset'] == dataset]
        diff = subset.filter(like='_diff').values.flatten()
        t_stat, p_value, result = t_test_mean_greater_than_zero(diff)
        results.append({'Category': 'Dataset', 'Name': dataset, 'T_Statistic': t_stat, 'P-value': p_value, 'Result': result})
    return pd.DataFrame(results)

def test_by_algorithm(df):
    """Performs t-tests for each algorithm to check if the mean difference is greater than zero."""
    results = []
    algorithms = df.columns[df.columns.str.contains('_diff')].str.replace('_diff', '', regex=False)
    for algorithm in algorithms:
        subset = df[f'{algorithm}_diff'].dropna()  # Drop NaN values
        t_stat, p_value, result = t_test_mean_greater_than_zero(subset)
        results.append({'Category': 'Algorithm', 'Name': algorithm, 'T_Statistic': t_stat, 'P-value': p_value, 'Result': result})
    return pd.DataFrame(results)

def test_by_type_overall(df):
    """Performs t-tests to check if the overall mean differences for each type are greater than zero."""
    results = []
    for type_ in df['Type'].unique():
        subset = df[df['Type'] == type_]
        diff = subset.filter(like='_diff').values.flatten()
        t_stat, p_value, result = t_test_mean_greater_than_zero(diff)
        results.append({'Category': 'Type (Overall)', 'Name': type_, 'T_Statistic': t_stat, 'P-value': p_value, 'Result': result})
    return pd.DataFrame(results)


def test_by_type_by_algorithm(df):
    """Performs t-tests to check if the mean differences for each type by algorithm are greater than zero."""
    results = []
    algorithms = df.columns[df.columns.str.contains('_diff')].str.replace('_diff', '', regex=False)
    for type_ in df['Type'].unique():
        for algorithm in algorithms:
            subset = df[(df['Type'] == type_)][f'{algorithm}_diff'].dropna()
            if not subset.empty:
                t_stat, p_value, result = t_test_mean_greater_than_zero(subset)
                results.append({'Category': 'Type (By Algorithm)', 'Type':type_, 'Alogorithm':algorithm, 'T_Statistic': t_stat, 'P-value': p_value, 'Result': result})
    return pd.DataFrame(results)
```

##Mannwhitney

```python

def mannwhitney_greater_than_zero(data):
    """Performs a one-sample Mann-Whitney U test to check if the median is greater than zero."""
    # Create a comparison array of zeros (the null hypothesis: median = 0)
    comparison = [0] * len(data)
    u_stat, p_value = mannwhitneyu(data, comparison, alternative='greater')
    result = "yes" if p_value < 0.05 else "no"
    return u_stat, p_value, result


def m_test_by_dataset(df):
    """Performs Mann-Whitney U tests for each dataset to check if the median difference is greater than zero."""
    results = []
    for dataset in df['Dataset'].unique():
        subset = df[df['Dataset'] == dataset]
        diff = subset.filter(like='_diff').values.flatten()
        if len(diff) > 1:  # Mann-Whitney requires at least two data points
            u_stat, p_value, result = mannwhitney_greater_than_zero(diff)
            results.append({'Category': 'Dataset', 'Name': dataset, 'U_Statistic': u_stat, 'P-value': p_value, 'Result': result})
    return pd.DataFrame(results)


def m_test_by_algorithm(df):
    """Performs Mann-Whitney U tests for each algorithm to check if the median difference is greater than zero."""
    results = []
    algorithms = df.columns[df.columns.str.contains('_diff')].str.replace('_diff', '', regex=False)
    for algorithm in algorithms:
        subset = df[f'{algorithm}_diff'].dropna()  # Drop NaN values
        if len(subset) > 1:  # Mann-Whitney requires at least two data points
            u_stat, p_value, result = mannwhitney_greater_than_zero(subset)
            results.append({'Category': 'Algorithm', 'Name': algorithm, 'U_Statistic': u_stat, 'P-value': p_value, 'Result': result})
    return pd.DataFrame(results)


def m_test_by_type_overall(df):
    """Performs Mann-Whitney U tests to check if the overall median differences for each type are greater than zero."""
    results = []
    for type_ in df['Type'].unique():
        subset = df[df['Type'] == type_]
        diff = subset.filter(like='_diff').values.flatten()
        if len(diff) > 1:  # Mann-Whitney requires at least two data points
            u_stat, p_value, result = mannwhitney_greater_than_zero(diff)
            results.append({'Category': 'Type (Overall)', 'Name': type_, 'U_Statistic': u_stat, 'P-value': p_value, 'Result': result})
    return pd.DataFrame(results)


def m_test_by_type_by_algorithm(df):
    """Performs Mann-Whitney U tests to check if the median differences for each type by algorithm are greater than zero."""
    results = []
    algorithms = df.columns[df.columns.str.contains('_diff')].str.replace('_diff', '', regex=False)
    for type_ in df['Type'].unique():
        for algorithm in algorithms:
            subset = df[(df['Type'] == type_)][f'{algorithm}_diff'].dropna()
            if len(subset) > 1:  # Mann-Whitney requires at least two data points
                u_stat, p_value, result = mannwhitney_greater_than_zero(subset)
                results.append({'Category': 'Type (By Algorithm)', 'Type': type_, 'Algorithm': algorithm, 'U_Statistic': u_stat, 'P-value': p_value, 'Result': result})
    return pd.DataFrame(results)
```

```python
from scipy.stats import shapiro
import pandas as pd
import numpy as np

def test_normality_by_group(df, level='dataset', alpha=0.05):
    """
    ทดสอบความเป็น normality โดยใช้ Shapiro-Wilk แยกตาม Dataset หรือ Type
    :param df: DataFrame ที่มี column 'Dataset', 'Type', และ *_diff
    :param level: 'dataset' หรือ 'type'
    :param alpha: ค่าความเชื่อมั่นของการทดสอบ (ค่า default = 0.05)
    :return: DataFrame ที่แสดงผลการทดสอบแต่ละกลุ่ม
    """
    results = []

    if level == 'dataset':
        groups = df['Dataset'].unique()
        group_column = 'Dataset'
        category = 'Dataset'
    elif level == 'type':
        groups = df['Type'].unique()
        group_column = 'Type'
        category = 'Type'
    else:
        raise ValueError("level ต้องเป็น 'dataset' หรือ 'type' เท่านั้น")

    for group in groups:
        subset = df[df[group_column] == group]
        diff_values = subset.filter(like='_diff').values.flatten()

        try:
            # Clean numeric only
            data = pd.Series(diff_values)
            data = pd.to_numeric(data, errors='coerce').dropna()
            data = np.array(data).astype(float).flatten()

            if len(data) < 3:
                result = 'too_few_data'
                stat, p_value = None, None
            else:
                stat, p_value = shapiro(data)
                result = 'non-normal' if p_value < alpha else 'normal'
        except Exception as e:
            stat, p_value, result = None, None, f'error: {str(e)}'

        results.append({
            'Category': category,
            'Name': group,
            'N': len(data),
            'Statistic': stat,
            'P-Value': p_value,
            'Result': result
        })

    return pd.DataFrame(results)
```
